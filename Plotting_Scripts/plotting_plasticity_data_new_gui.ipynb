```python
import numpy as np
import matplotlib.pyplot as plt

# Sample data (Replace this with your actual data)
experiment_to_plot = sucrose_plasticity_traces_no_delay

# Initialize lists to store combined data
test_pathway_times_baseline = []
test_pathway_time_after_LTP = []
test_pathway_peaks_baseline = []
test_pathway_peaks_after_LTP = []

control_pathway_times_baseline = []
control_pathway_time_after_LTP = []
control_pathway_peaks_baseline = []
control_pathway_peaks_after_LTP = []

# Collect data across all cells
for cell_name, cell_data in experiment_to_plot.items():
    for channel, baseline_peaks in cell_data['max_peaks_baseline'].items():
        after_LTP_peaks = cell_data['max_peaks_after_LTP'][channel]
        baseline_times = np.array(cell_data['baseline_time_points'])
        after_LTP_times = np.array(cell_data['after_LTP_time_points'])
        
        # Normalize baseline peaks
        baseline_mean = np.mean(baseline_peaks)
        normalized_baseline_peaks = baseline_peaks / baseline_mean
        
        # Normalize after LTP peaks relative to the normalized baseline
        normalized_after_LTP_peaks = np.array([after_LTP_peak / baseline_mean for after_LTP_peak in after_LTP_peaks])
        
        # Zero the time points
        baseline_times_zeroed = baseline_times - baseline_times[0]
        after_LTP_times_zeroed = after_LTP_times - baseline_times[0]
        
        # Exclude baseline data points beyond 10 minutes
        baseline_mask = baseline_times_zeroed <= 10
        baseline_times_zeroed_filtered = baseline_times_zeroed[baseline_mask]
        normalized_baseline_peaks_filtered = normalized_baseline_peaks[:len(baseline_times_zeroed_filtered)]
        
        # Exclude after LTP data points beyond 40 minutes
        after_LTP_mask = after_LTP_times_zeroed <= 40
        after_LTP_times_zeroed_filtered = after_LTP_times_zeroed[after_LTP_mask]
        normalized_after_LTP_peaks_filtered = normalized_after_LTP_peaks[after_LTP_mask]
        
        # Separate data into Test Pathway and Control Pathway
        if 'channel_1' in channel:  # Test Pathway
            test_pathway_times_baseline.extend(baseline_times_zeroed_filtered)
            test_pathway_peaks_baseline.extend(normalized_baseline_peaks_filtered)
            test_pathway_time_after_LTP.extend(after_LTP_times_zeroed_filtered)
            test_pathway_peaks_after_LTP.extend(normalized_after_LTP_peaks_filtered)
        elif 'channel_2' in channel:  # Control Pathway
            control_pathway_times_baseline.extend(baseline_times_zeroed_filtered)
            control_pathway_peaks_baseline.extend(normalized_baseline_peaks_filtered)
            control_pathway_time_after_LTP.extend(after_LTP_times_zeroed_filtered)
            control_pathway_peaks_after_LTP.extend(normalized_after_LTP_peaks_filtered)

# Convert lists to numpy arrays
test_pathway_times_baseline = np.array(test_pathway_times_baseline)
test_pathway_peaks_baseline = np.array(test_pathway_peaks_baseline)
test_pathway_time_after_LTP = np.array(test_pathway_time_after_LTP)
test_pathway_peaks_after_LTP = np.array(test_pathway_peaks_after_LTP)

control_pathway_times_baseline = np.array(control_pathway_times_baseline)
control_pathway_peaks_baseline = np.array(control_pathway_peaks_baseline)
control_pathway_time_after_LTP = np.array(control_pathway_time_after_LTP)
control_pathway_peaks_after_LTP = np.array(control_pathway_peaks_after_LTP)

# Define the function to calculate means and standard errors
def calculate_mean_and_std_error(times, values, bin_size=1):
    time_bins = np.arange(0, np.max(times) + bin_size, bin_size)
    means = np.zeros(len(time_bins) - 1)
    std_errors = np.zeros(len(time_bins) - 1)
    
    for i in range(len(time_bins) - 1):
        mask = (times >= time_bins[i]) & (times < time_bins[i+1])
        bin_values = values[mask]
        if len(bin_values) > 0:
            means[i] = np.mean(bin_values)
            std_errors[i] = np.std(bin_values) / np.sqrt(len(bin_values))
        else:
            means[i] = np.nan
            std_errors[i] = np.nan
    
    return time_bins[:-1], means, std_errors

# Calculate means and standard errors for baseline
test_pathway_time_bins_baseline, test_pathway_means_baseline, test_pathway_std_errors_baseline = calculate_mean_and_std_error(test_pathway_times_baseline, test_pathway_peaks_baseline)
control_pathway_time_bins_baseline, control_pathway_means_baseline, control_pathway_std_errors_baseline = calculate_mean_and_std_error(control_pathway_times_baseline, control_pathway_peaks_baseline)

# Calculate means and standard errors for after LTP
test_pathway_time_bins_after_LTP, test_pathway_means_after_LTP, test_pathway_std_errors_after_LTP = calculate_mean_and_std_error(test_pathway_time_after_LTP, test_pathway_peaks_after_LTP)
control_pathway_time_bins_after_LTP, control_pathway_means_after_LTP, control_pathway_std_errors_after_LTP = calculate_mean_and_std_error(control_pathway_time_after_LTP, control_pathway_peaks_after_LTP)

# Initialize figure and axes
fig, axs = plt.subplots(2, 1, figsize=(12, 10))

# Plot for Test Pathway
axs[0].errorbar(test_pathway_time_bins_baseline, test_pathway_means_baseline, yerr=test_pathway_std_errors_baseline, fmt='o', color='black') 
axs[0].errorbar(test_pathway_time_bins_after_LTP, test_pathway_means_after_LTP, yerr=test_pathway_std_errors_after_LTP, fmt='o', color='black') 
axs[0].axvline(x=10, color='r', linestyle='--', label='Start of LTP Induction')
axs[0].set_title('No Delay Sucrose - Test Pathway')
axs[0].set_xlabel('Time (min)')
axs[0].set_ylabel('Normalized Max Peak')
axs[0].set_ylim(0, 5)
axs[0].legend()

# Plot for Control Pathway
axs[1].errorbar(control_pathway_time_bins_baseline, control_pathway_means_baseline, yerr=control_pathway_std_errors_baseline, fmt='o', color='black')
axs[1].errorbar(control_pathway_time_bins_after_LTP, control_pathway_means_after_LTP, yerr=control_pathway_std_errors_after_LTP, fmt='o', color='black')

axs[1].axvline(x=10, color='r', linestyle='--', label='Start of LTP Induction')
axs[1].set_title('Control Pathway')
axs[1].set_xlabel('Time (min)')
axs[1].set_ylabel('Normalized Max Peak')
axs[1].set_ylim(0, 5)
axs[1].legend()

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

num_cells = len(GNB1_plasticity_traces)
channel_labels = ['Test Pathway', 'Control Pathway']
colors = ['black', 'black']  # Colors for the scatter plot (all points in black)

# Adjust subplot creation based on the number of cells and channels
fig, axs = plt.subplots(num_cells, 2, figsize=(12, 6*num_cells), sharex=True, sharey=True)

# Ensure axs is always 2-dimensional for consistent indexing
if num_cells == 1:
    axs = np.expand_dims(axs, axis=0)
if len(channel_labels) == 1:
    axs = np.expand_dims(axs, axis=1)

fig.suptitle('Normalized Max Peaks for GNB1 Plasticity Traces', fontsize=16)

for i, (cell_name, cell_data) in enumerate(GNB1_plasticity_traces.items()):
    for j, (channel, baseline_peaks) in enumerate(cell_data['max_peaks_baseline'].items()):
        after_LTP_peaks = cell_data['max_peaks_after_LTP'][channel]
        baseline_times = np.array(cell_data['baseline_time_points'])
        after_LTP_times = np.array(cell_data['after_LTP_time_points'])
        
        # Normalize time points to start from zero
        baseline_times_zeroed = baseline_times - baseline_times[0]
        after_LTP_times_zeroed = after_LTP_times - baseline_times[0]
        
        # Filter out times greater than 10 minutes for baseline and 40 minutes for after LTP
        baseline_mask = baseline_times_zeroed <= 10
        after_LTP_mask = after_LTP_times_zeroed <= 40
        
        filtered_baseline_times = baseline_times_zeroed[baseline_mask]
        filtered_baseline_peaks = np.array(baseline_peaks)[baseline_mask]
        
        filtered_after_LTP_times = after_LTP_times_zeroed[after_LTP_mask]
        filtered_after_LTP_peaks = np.array(after_LTP_peaks)[after_LTP_mask]
        
        # Normalize baseline peaks using the first few baseline points
        normalization_window = 5  # Number of initial points to consider for normalization
        if len(filtered_baseline_peaks) >= normalization_window:
            baseline_normalization_value = np.mean(filtered_baseline_peaks[:normalization_window])
        else:
            baseline_normalization_value = np.mean(filtered_baseline_peaks)
        
        print(f"Cell: {cell_name}, Channel: {channel}, Baseline Normalization Value: {baseline_normalization_value}")
        print(f"First few baseline peaks: {filtered_baseline_peaks[:normalization_window]}")
        print(f"First few after LTP peaks: {filtered_after_LTP_peaks[:normalization_window]}")
        
        normalized_baseline_peaks = filtered_baseline_peaks / baseline_normalization_value
        normalized_after_LTP_peaks = filtered_after_LTP_peaks / baseline_normalization_value
        
        # Plot for Test Pathway and Control Pathway on separate subplots
        ax = axs[i, j]
        ax.scatter(filtered_baseline_times, normalized_baseline_peaks, color=colors[j], label=f'{channel_labels[j]} Baseline Peaks', marker='o')
        ax.scatter(filtered_after_LTP_times, normalized_after_LTP_peaks, color=colors[j], marker='x', label=f'{channel_labels[j]} After LTP Peaks')
        ax.axvline(x=10, color='r', linestyle='--', label='Start of LTP Induction')
        ax.set_title(f"{cell_name} - {channel_labels[j]}")
        ax.set_xlabel('Time (min)')
        ax.set_ylabel('Normalized Max Peak')
        ax.set_ylim(-1, 5)
        ax.legend()

plt.tight_layout()
plt.show()
```
