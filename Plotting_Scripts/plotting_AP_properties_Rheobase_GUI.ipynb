### WT Properties stats

```python
p_values = {}
for key in AP_properties_to_plot:
    data_NMDG_older = clean_data(AP_properties_dict_older[key])
    data_NMDG = clean_data(AP_properties_dict_NMDG[key])
    data_sucrose = clean_data(AP_properties_dict_sucrose_GNB1[key])

    # Perform t-tests and store p-values
    p_values[f'{key}_NMDG_older_vs_NMDG'] = stats.ttest_ind(data_NMDG_older, data_NMDG, equal_var=False)[1]
    p_values[f'{key}_NMDG_older_vs_sucrose'] = stats.ttest_ind(data_NMDG_older, data_sucrose, equal_var=False)[1]
    p_values[f'{key}_NMDG_vs_sucrose'] = stats.ttest_ind(data_NMDG, data_sucrose, equal_var=False)[1]

# Flatten the list of p-values
p_values_list = list(p_values.values())

# Correct the p-values using the Bonferroni correction
p_values_corrected = multipletests(p_values_list, method='bonferroni')[1]

# Map corrected p-values back to the original structure
p_values_corrected_dict = {}
for idx, key in enumerate(p_values.keys()):
    p_values_corrected_dict[key] = p_values_corrected[idx]
```

# Getting average traces at rheobase for all cells

```python
def calculate_decay_area(rheobase_traces, AP_thresholds, title, sampling_rate=20000, duration_ms=50, tau=20):
    """
    Calculate the decay area from the AHP trough to a fixed duration after, plot the trace, and return decay area and AP traces.

    Parameters:
        rheobase_traces (dict): Dictionary containing rheobase traces for each cell.
        AP_thresholds (dict): Dictionary containing AP thresholds for each cell.
        sampling_rate (int): Sampling rate of the traces (in Hz).
        duration_ms (int): Duration after the AHP trough to consider (in milliseconds).
        tau (float): Time constant for the exponential decay (in milliseconds).

    Returns:
        decay_areas (list): List of decay areas for each trace.
        traces (list): List of processed AP traces.
    """
    dt = 1 / sampling_rate  # Time step based on sampling rate
    duration_samples = int(duration_ms / 1000 / dt)  # Convert duration to number of samples

    decay_areas = []
    traces = []

    fig, ax = plt.subplots(1, 1, figsize=(12, 5))

    for cell in rheobase_traces:
        if cell in AP_thresholds:
            for idx, trace in enumerate(rheobase_traces[cell]['Rheobase_Trace']):
                current_trace = rheobase_traces[cell]['Rheobase_Trace'][idx]
                current_AP_threshold = AP_thresholds[cell]['AP_threshold']

                # Check if current_AP_threshold is a list of lists
                if isinstance(current_AP_threshold[0], list):
                    AP_threshold_value = current_AP_threshold[0][0]
                else:
                    AP_threshold_value = current_AP_threshold[0]

                trace_copy = current_trace.copy()

                # Find the peak index of the first spike
                peak_idx = np.argmax(trace_copy)

                # Find the AHP trough (minimum value) after the first spike peak
                AHP_trough_idx = peak_idx + np.argmin(trace_copy[peak_idx:])

                # Define the end index for a fixed 50 ms after the AHP trough
                end_idx = AHP_trough_idx + duration_samples
                
                amplitude = AP_threshold_value - trace_copy[AHP_trough_idx]  # Amplitude at the AHP trough
                
                

                # Ensure that the end index does not exceed the length of the trace
                if end_idx >= len(trace_copy):
                    end_idx = len(trace_copy) - 1

                # Calculate the area under the trace from the AHP trough to 50 ms after
                clipped_trace_area_AHP = AP_threshold_value - trace_copy[AHP_trough_idx:end_idx + 1]
                trace_area_AHP = clipped_trace_area_AHP/amplitude 
                initial_area = np.trapz(trace_area_AHP, dx=dt)  # Initial area under the trace is weighted tau
                
                decay_area = initial_area * 1000 # Convert to ms

                decay_areas.append(decay_area)
                traces.append(current_trace)

                print(f"Decay area for Cell {cell}, Trace {idx}: {decay_area} ms")

                # Generate current_time array based on the sampling rate and length of the trace
                current_time = np.arange(0, len(current_trace) * dt, dt) 

                ax.plot(current_time, current_trace, color='black', alpha=0.1)

                # Fill the area under the trace from the AHP trough to 50 ms after
                ax.fill_between(current_time[AHP_trough_idx:end_idx + 1], clipped_trace_area_AHP, trace_copy[AHP_trough_idx], color='green', alpha=0.3)

                # Optional: Plot the AHP trough and the end point of the 50 ms interval
                ax.scatter(current_time[AHP_trough_idx], trace_copy[AHP_trough_idx], color='blue')
                ax.scatter(current_time[end_idx], trace_copy[end_idx], color='orange')

                # Break after the first AHP trough is processed to ensure only the first is used
                break

    plt.xlabel('Time (s)')
    plt.ylabel('Voltage (mV)')
    plt.title(f'AP Trace with Decay Area {title}') 
    plt.show()

    return decay_areas, traces
```

### Plot an Example WT and GNB1 Rheobase Trace

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from scipy.signal import find_peaks

data_file = '/Users/samgritz/Library/CloudStorage/GoogleDrive-gritz122@gmail.com/My Drive/Rutgers/Milstein_Lab_Data/GNB1_WT_Project/New_Analysis_Processed_Data/WT_data/Processed_data/WT_042024_on_processed_data/Sucrose/07172024_c3_processed_new.pkl'

# Load the data file
data_df = pd.read_pickle(data_file)
sweep_list = [31]  
AP_trace_end = 50  # Duration after the AP threshold to consider for the AHP

# Extract cell name and date from the file name
file_name = data_file.split('/')[-1].replace('.pkl', '')
cell_date, cell_name = file_name.split('_')[:2]
data_name = cell_name + '_' + cell_date

# Initialize dictionary to store traces
rheobase_traces = {}

# Process the sweeps
for sweep in sweep_list:
    current_analysis_data = data_df['analysis_dict'][sweep]
    
    # Get the trace and find its length
    current_rheobase_trace = data_df['sweep'][sweep].copy()  # Ensure it's a writable copy
    
    # Make a list of traces that are 4 sweeps after the rheobase trace
    additional_traces = []
    for i in range(1, 3):
        if sweep + i < len(data_df['sweep']):
            additional_trace = data_df['sweep'][sweep + i].copy()
            additional_traces.append(additional_trace)
    
    trace_length = len(current_rheobase_trace)

    # Compute the time vector
    time = np.arange(0, trace_length * 0.05, 0.05)

    # Compute the first derivative
    derivative = np.diff(current_rheobase_trace)

    # Find AP Threshold as the peak of the first derivative
    ap_threshold_index = np.argmax(derivative)

    # Find AHP index as the argmin after the spike
    end_index = ap_threshold_index + int(AP_trace_end * 20000 / 1000)
    if end_index > trace_length:
        end_index = trace_length
    post_spike_trace = current_rheobase_trace[ap_threshold_index:end_index]
    ahp_index = np.argmin(post_spike_trace) + ap_threshold_index

    # Find the peak of the first spike
    peaks, _ = find_peaks(current_rheobase_trace)
    if len(peaks) > 0:
        # Get the index of the first peak
        first_spike_peak_index = peaks[0]
    else:
        first_spike_peak_index = None

    # Store the traces
    if data_name not in rheobase_traces:
        rheobase_traces[data_name] = []

    rheobase_traces[data_name].append((current_rheobase_trace, time, ap_threshold_index, ahp_index, first_spike_peak_index, additional_traces))

# Plotting
for data_name, traces in rheobase_traces.items():
    plt.figure(figsize=(12, 8))

    for trace, time, ap_threshold_index, ahp_index, first_spike_peak_index, additional_traces in traces:
        # Plot rheobase trace
        plt.plot(time, trace, color='red', label='Rheobase Trace')

        # Scatter for AP Threshold
        if ap_threshold_index < len(trace):
            plt.scatter(time[ap_threshold_index], trace[ap_threshold_index], color='blue', label='AP Threshold')

        # Scatter for AHP and Spike Peak
        if ahp_index < len(trace):
            plt.scatter(time[ahp_index], trace[ahp_index], color='green', label='AHP')
        if first_spike_peak_index is not None and first_spike_peak_index < len(trace):
            plt.scatter(time[first_spike_peak_index], trace[first_spike_peak_index], color='red', label='Spike Peak')
        
        # Plot additional sweeps in grey
        for additional_trace in additional_traces:
            additional_time = np.arange(0, len(additional_trace) * 0.05, 0.05)

            plt.plot(additional_time, additional_trace, color='grey', alpha=0.5, label='Additional Sweeps')
    
    plt.xlabel('Time (ms)')
    plt.ylabel('Voltage (mV)')
    plt.xlim(300,900) 
    plt.title(f'Rheobase and Additional Sweeps for {data_name}')
    plt.legend()
    plt.grid(False)  # Remove grid
    plt.show()
    
    #save the figure as a png and svg
    plt.savefig('/Users/samgritz/Library/CloudStorage/GoogleDrive-gritz122@gmail.com/My Drive/Rutgers/Milstein_Lab_Data/GNB1_WT_Project/New_Analysis_Processed_Data/GNB1_data/Processed_data/saved_figures/GNB1_vs_WT_data/AP_properties_FI_example.svg')
```
