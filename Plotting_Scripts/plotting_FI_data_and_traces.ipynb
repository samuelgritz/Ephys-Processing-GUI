```python
# Convert dictionaries to DataFrame
def dict_to_df(data_dict, genotype_label, holding_potential_dict, slicing_solution):
    data = []
    for cell, currents in data_dict.items():
        for current, firing_rate in currents.items():
            if cell in holding_potential_dict:
                holding_potential = holding_potential_dict[cell]
                data.append({'Cell': cell, 'Current_Amplitude': current, 'Firing_Rate': firing_rate, 'Genotype': genotype_label, 'Holding_Potential': holding_potential, 'Slicing_Solution': slicing_solution}) 
    return pd.DataFrame(data)
```

```python
def plot_ISI_times(ISI_data, title): 
    fig, ax = plt.subplots(1, 1, figsize=(6, 5))
    AP_numbers_to_plot = {}
    sorted_ISI_times_to_plot = {}
    num_cells = len(ISI_data)
    fig.suptitle(f' N = {num_cells}') 

    for cell in ISI_data: 
        if cell not in AP_numbers_to_plot:
            AP_numbers_to_plot[cell] = []
            sorted_ISI_times_to_plot[cell] = []

        for current_amplitude_data in ISI_data[cell]: 
            for AP_number in sorted(ISI_data[cell][current_amplitude_data].keys()):
                if AP_number <= 7:
                    current_AP_number = AP_number
                    AP_numbers_to_plot[cell].append(current_AP_number)  
                    current_sorted_ISI_time = ISI_data[cell][current_amplitude_data][AP_number]
                    sorted_ISI_times_to_plot[cell].append(current_sorted_ISI_time) 

    for cell in AP_numbers_to_plot:
        if cell in sorted_ISI_times_to_plot.keys():
            current_cell = cell
            plotting_AP_numbers = AP_numbers_to_plot[cell][:7]
            plotting_ISI_times = sorted_ISI_times_to_plot[cell][:7]



            ax.plot(plotting_AP_numbers, plotting_ISI_times, label = f'{current_cell}')
            ax.set_xticks(np.arange(2, 8, 1) ) 
            ax.set_xticklabels(['1', '2' , '3' , '4', '5', '6']) 
            ax.set_title(title)
            ax.set_ylabel('ISI Time (ms)') 
            ax.set_xlabel('AP Spike Number')
            ax.legend()

#plot_ISI_times(ISI_data_NMDG, 'ISI times NMDG WT') 
#plot_ISI_times(ISI_data_sucrose, 'ISI times Sucrose WT') 
#plot_ISI_times(ISI_data_older, 'ISI times Older WT Data') 

#plot_ISI_times(ISI_data_NMDG_GNB1, 'ISI times NMDG GNB1') 
#plot_ISI_times(ISI_data_sucrose_GNB1, 'ISI times Sucrose GNB1') 
# plot_ISI_times(ISI_data_older_GNB1, 'ISI times Older GNB1')  
```

```python
import pandas as pd

def dict_to_df_ISI(data_dict, genotype_label):
    data = []
    for cell, current_amplitudes in data_dict.items():
        for current_amplitude, isi_data in current_amplitudes.items():
            for ap_number, isi_time in isi_data.items():
                data.append({'Cell': cell, 
                             'Current_Amplitude': current_amplitude, 
                             'AP_Number': ap_number, 
                             'ISI_Time': isi_time, 
                             'Genotype': genotype_label})
    return pd.DataFrame(data, columns=['Cell', 'Current_Amplitude', 'AP_Number', 'ISI_Time', 'Genotype'])
```

```python
def get_FI_trace_example(dir_path):
# Initialize a dictionary to store traces by cell name
    dir_path = dir_path
    example_traces = []
    current_amplitudes = []
    data_file = pd.read_pickle(dir_path)

    # Collect current amplitudes and traces
    for i in range(len(data_file)):
        analysis_data = data_file['analysis_dict'][i]

        if 'Coarse_FI' in analysis_data.keys():
            trace = data_file['sweep'][i]
            current_amplitude = analysis_data['Coarse_FI']['current_amplitudes']
            example_traces.append(trace)
            current_amplitudes.append(current_amplitude)
            
        
        elif 'IV_stim' in analysis_data.keys():
            trace = data_file['sweep'][i]
            current_amplitude = analysis_data['IV_stim']['current_amplitudes']
            example_traces.append(trace)
            current_amplitudes.append(current_amplitude)
            
            
    return example_traces, current_amplitudes
```
